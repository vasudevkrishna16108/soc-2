```python
# unsupervised learning algorithms
# k means clustering
# k means clustering is an unsupervised learning algorithm that will attempt to group similar clusters together from your data 
# It is mainly used in clustering similar documents 
# clustering customers based on similar features
```

```python
from sklearn.datasets import make_blobs 
# used to generate synthetic dataset
# for clustering and classification tasks
# this function will create clusters of data points with 
# gaussiam distribution
```

```python
# creating random dataset
data = make_blobs(n_samples=200,n_features=2,centers=3,cluster_std=5.6,random_state=101)
# n_samples = total no of points equally divided among clusters.
# n_features = It indicates the no of features(columns)
# centers = It determines no of clusters to be generated
# cluster_std = It sets the standards deviation of the clusters , high value makes the clusters to spread out
data
```

```python
import matplotlib.pyplot as plt 
x , y = data
plt.scatter(x[:, 0],x[:,1],c=y, cmap = 'rainbow',edgecolor='black',s=100)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Scatter plot')
```

```python
data[0].shape
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(10,6))
ax1.set_title('K means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')
ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
df_feat = pd.DataFrame(scalar_standard,columns=df.columns[:-1])
df_feat
```

```python
df_feat.isnull().sum()
```

```python
df_feat.isna().sum()
```

```python
from sklearn.model_selection import train_test_split
x = df_feat
y = df['TARGET CLASS']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state=101)
x_train.shape
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=13)
knn
```

```python
# To train the model
knn.fit(x_train,y_train)
```

```python
pred = knn.predict(x_test)
pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score = accuracy_score(pred,y_test)
accuracy_score
```

```python
error_rate = []
for sal in range(1,30):
    knn = KNeighborsClassifier(n_neighbors=sal)
    knn.fit(x_train,y_train)
    pred_i = knn.predict(x_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
error_rate
```

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(6,3))
plt.plot(range(1,30),error_rate)
plt.title("Error rate vs value")
plt.xlabel("K Value")
plt.ylabel("Error Rate")
```
